````python
## import library
import pandas as pd
````

````python
## Ingest remote data
df = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/ppp_excerpt.csv")
df
````

````python
## get info about data types
df.info()
````

### ```.agg()```

- Use the ```.agg()``` that allows you call specific summary statistics at one time.

````python
df["CurrentApprovalAmount"].count()
````

````python
df["CurrentApprovalAmount"].mean()
````

````python
df["CurrentApprovalAmount"].sum()
````

## ```groupby```

````python
## What is the sum, mean and count for the current 
## approval amount for owners by race 
race_gb = df.groupby("Race")
race_gb
race_gb.describe()
````

````python
race_gb["CurrentApprovalAmount"].sum().to_frame()
````

````python
race_gb["CurrentApprovalAmount"].count().to_frame()
````

````python
race_gb["CurrentApprovalAmount"].mean().to_frame()
````
